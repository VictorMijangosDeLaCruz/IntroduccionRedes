## Datasets no estructurados

Los datasets no estructurados responden a problemas más específicos. Por su misma naturaleza, existe menos variedad de bases de datos sobre este tipo de datasets, pero se pueden encontrar algunos repositorios para imágenes y textos:

### Imágenes

Algunos repositorios para imágenes son:

1. Celeb A: http://mmlab.ie.cuhk.edu.hk/projects/CelebA.html
2. Yale Face Database: http://vision.ucsd.edu/content/yale-face-database
3. LabelMe: http://labelme.csail.mit.edu/Release3.0/
4. Open Image Dataset: https://storage.googleapis.com/openimages/web/index.html
5. MNIST: http://yann.lecun.com/exdb/mnist/

### Texto

Algunos repositorios con datos textuales son:

1. HuggingFace: https://huggingface.co/datasets
2. IMDB Datasets: https://developer.imdb.com/non-commercial-datasets/
3. Stanford Sentiment Treebank: https://nlp.stanford.edu/sentiment/code.html


Dentro de Pytorch podemos encontrar también datasets para estos tipos de datos, en particular, en los módulos:

1. TorchVision
2. TorchText

```python
import torch
import math
import torchvision
import matplotlib.pyplot as plt
from torch import nn
```

```python
#OxfordIIITPet dataset de imágenes de mascotas
data = torchvision.datasets.OxfordIIITPet(root=".", download=True)
```

```python
def show_images(datset, num_samples=6, cols=2):
    """ Función para visualizar imágenes del dataset """
    plt.figure(figsize=(20,15)) 
    for i, img in enumerate(datset):
        if i == num_samples:
            break
        plt.subplot(num_samples//cols + 1, cols, i + 1)
        plt.imshow(img[0])
        
show_images(data)
```
